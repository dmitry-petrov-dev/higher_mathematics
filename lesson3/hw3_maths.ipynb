## 3. Задание (в программе)

Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# окружность
t = np.linspace(0, 2*np.pi,100)
x0 = 0
y0 = 0
plt.axes().set_aspect(1)
plt.plot(x0 + np.cos(t), y0 + np.sin(t))
```

```python
# эллипс
t = np.linspace(0, 2*np.pi,100)
x0 = 0
y0 = 0
a = 2
b = 1
plt.axes().set_aspect(1)
plt.plot(x0 + a*np.cos(t),y0 + b*np.sin(t))
```

```python
# гипербола
t = np.linspace(-np.pi/3, np.pi/3,100)
plt.plot(np.tan(t), 1/np.cos(t))
t = np.linspace(np.pi-np.pi/3, np.pi+np.pi/3,200)
plt.plot(np.tan(t), 1/np.cos(t))
```

```python
# трехмерный график двух параллельных плоскостей
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
X, Y = np.meshgrid(X, Y)
Z1 = X + Y
Z2 = X + Y - 20
ax.plot_surface(X, Y, Z1, color='green')
ax.plot_surface(X, Y, Z2, color='blue')
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```python
# трехмерный график двух любых поверхностей второго порядка
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
X, Y = np.meshgrid(X, Y)
Z1 = X**2 + Y**2
Z2 = X**2 - 3*Y**2 - 20
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2, color='yellow')
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1)/x)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,8) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1-y) - 1)

x1, y1 =  fsolve(equations, (-2, 1))
              
print (x1, y1)
```
